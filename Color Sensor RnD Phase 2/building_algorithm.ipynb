```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
def normalize(df):
    
    row_sums = df.sum(axis = 1)

    normalized_df = df.div(row_sums, axis=0)

    return normalized_df
```

```python
bblue = normalize(pd.read_csv("bblue.csv"))
bgreen = normalize(pd.read_csv("bgreen.csv"))
borange = normalize(pd.read_csv("borange.csv"))
bpurple = normalize(pd.read_csv("bpurple.csv"))
bred = normalize(pd.read_csv("bred.csv"))
byellow = normalize(pd.read_csv("byellow.csv"))
```

```python
# Remove the initialization row (first row) from each dataframe
bblue_clean = bblue.iloc[1:]
borange_clean = borange.iloc[1:]
byellow_clean = byellow.iloc[1:]
bpurple_clean = bpurple.iloc[1:]
# Now combine the cleaned data

bblue_clean.columns = ['R', 'G', 'B']
borange_clean.columns = ['R', 'G', 'B']
byellow_clean.columns = ['R', 'G', 'B']
bpurple_clean.columns = ['R', 'G', 'B']




X = pd.concat([bblue_clean, borange_clean, byellow_clean, bpurple_clean], ignore_index=True)



print("Shape of X:", X.shape)

# Create labels - make sure the number matches your actual data
# If each cleaned dataset has 199 rows now:
y= ['blue']*198 + ['orange']*198 + ['yellow']*198 + ['purple']*198
```

```python
clf = DecisionTreeClassifier()

clf.fit(X, y)
```

```python
plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()
```
